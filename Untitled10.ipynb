```python
venues_map = folium.Map(location=[latitude, longitude], zoom_start=13)

for lat, lng, label in zip(df_india['location.lat'], df_india['location.lng'], df_india['name']):
    folium.CircleMarker(
        [lat, lng],
        radius=5,
        color='blue',
        popup=label,
        fill = True,
        fill_color='blue',
        fill_opacity=0.6
    ).add_to(venues_map)


for lat, lng, label in zip(df['location.lat'], df['location.lng'], df['name']):
    folium.CircleMarker(
        [lat, lng],
        radius=5,
        color='red',
        popup=label,
        fill = True,
        fill_color='blue',
        fill_opacity=0.6
    ).add_to(venues_map)

       

# display map
venues_map
```
